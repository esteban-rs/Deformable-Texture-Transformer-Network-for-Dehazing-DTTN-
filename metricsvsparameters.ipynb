```python
import matplotlib.pyplot as plt
```

```python
# Data from the plot
models     = ['GridDehazeNet', 'AECR-Net', 'MAXIM', 'PMNet', 'DehazeFormer-L', 'FFA-Net', 'Ours']
parameters = [1,                     2.61,   14.10,    18.9,             14.4,     4.456,  7.56]  # in millions (M)
psnr       = [32.16,                 37.2,   38.11,   38.41,            34.29,     36.39, 36.59]  # in dB
ssmi       = [.984,                  0.99,   0.991,   0.990,            0.984,     0.989, 0.992]  # in dB
```

```python
marker_styles = {
    'GridDehazeNet': ('s', 'purple'),
    'AECR-Net': ('p', 'green'),
    'MAXIM': ('h', 'black'),
    'PMNet': ('D', 'cyan'),
    'DehazeFormer-L': ('o', 'blue'),
    'FFA-Net': ('v', 'yellow'),
    'Ours': ('*', 'red')
}
```

```python
for model, x, y in zip(models, parameters, psnr):
    marker, color = marker_styles[model]
    size = 200 if model == 'Ours' else 100
    plt.scatter(x, y, marker=marker, color=color, s=size, label=model, edgecolors='k' if model != 'Ours' else 'none')

# Labels and legend
plt.xlabel('Parameters (M)')
plt.ylabel('PSNR (dB)')
plt.ylim(30, 40)
plt.legend()
plt.grid(True)

# Display plot
plt.tight_layout()
plt.savefig('model_comparison_psnr.png', dpi=300, bbox_inches='tight')  # Saves as high-quality PNG
plt.show()
```

```python
for model, x, y in zip(models, parameters, ssmi):
    marker, color = marker_styles[model]
    size = 200 if model == 'Ours' else 100
    plt.scatter(x, y, marker=marker, color=color, s=size, label=model, edgecolors='k' if model != 'Ours' else 'none')

# Labels and legend
plt.xlabel('Parameters (M)')
plt.ylabel('SSIM (dB)')
plt.ylim(0.98, 0.995)
plt.legend()
plt.grid(True)

# Display plot
plt.tight_layout()
plt.savefig('model_comparison_ssmi.png', dpi=300, bbox_inches='tight')  # Saves as high-quality PNG
plt.show()
```
