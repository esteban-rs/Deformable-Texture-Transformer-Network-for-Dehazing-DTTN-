```python
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
haze_dir   = r'/home/esteban/Data/RESIDE-IN/test/hazy/'
grid_dir   = r'/home/esteban/Data/RESIDE-IN/pred_GridDehazeNet_its/'
ffa_dir    = r'/home/esteban/Data/RESIDE-IN/pred_FFA_its/'
maxim_dir  = r'/home/esteban/Data/RESIDE-IN/pred_MAXIM_its/'
dttn_dir   = r'/home/esteban/Data/RESIDE-IN/pred_DTTN_its/'
dehaze_dir = r'/home/esteban/Data/RESIDE-IN/test/GT/'
```

```python
imgs_names = ['1405_10.png', '1413_10.png', '1401_10.png', '1415_10.png']
```

```python
1401,1415
```

```python
for img_name in imgs_names:
    images = [
        ('Hazy Input',     os.path.join(haze_dir, img_name)),
        ('GridDehazeNet',  os.path.join(grid_dir, img_name)),
        ('FFA-Net',        os.path.join(ffa_dir, img_name.replace('.', '_FFA.'))),
        ('MAXIM',          os.path.join(maxim_dir, img_name)),
        ('DTTN (Ours)',    os.path.join(dttn_dir, img_name.replace('.', '_DTTN.'))),
        ('Ground Truth',   os.path.join(dehaze_dir, img_name)),
    ]

    # Set up a 2x3 plot with tighter layout
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))
    
    # Minimize white space between plots
    fig.subplots_adjust(hspace=0.01, wspace=0.01, left=0.01, right=0.99, top=0.95, bottom=0.05)

    for i, ax in enumerate(axes.flat):
        title, path = images[i]
        if os.path.exists(path):
            img = Image.open(path).convert("RGB")
            
            ax.imshow(img)
            ax.set_title(title, fontsize=16)
        else:
            ax.text(0.5, 0.5, 'Image Not Found', fontsize=14, ha='center')
        ax.axis('off')

    # Save or show
    plt.savefig(f"./comparations/comparison_{img_name}", dpi=300, bbox_inches='tight', pad_inches=0)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
```

```python
images = [
        ('Hazy Input',     os.path.join(haze_dir, img_name)),
        ('GridDehazeNet',  os.path.join(grid_dir, img_name)),
        ('FFA-Net',        os.path.join(ffa_dir, img_name.replace('.', '_FFA.'))),
        ('MAXIM',          os.path.join(maxim_dir, img_name)),
        ('DTTN (Ours)',    os.path.join(dttn_dir, img_name.replace('.', '_DTTN.'))),
        ('Ground Truth',   os.path.join(dehaze_dir, img_name)),
    ]

fig, axes = plt.subplots(2, 3, figsize=(12, 8))
    
fig.subplots_adjust(hspace=0.01, wspace=0.01, left=0.01, right=0.99, top=0.95, bottom=0.05)

for i, ax in enumerate(axes.flat):
    title, path = images[i]
    if os.path.exists(path):
        img = Image.open(path).convert("RGB")
        ax.imshow(img)

        rects = [
            (40, 350, 200, 200),   # Left rectangle
            (480, 80, 120, 200)   # Right rectangle
        ]

        for (x, y, w, h) in rects:
            rect = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='red', facecolor='none')
            ax.add_patch(rect)
        ax.axis('off')
        ax.set_title(title, fontsize=16)
    else:
        ax.text(0.5, 0.5, 'Image Not Found', fontsize=14, ha='center')

    


# Save or show
plt.savefig(f"./comparations/comparison_local_{img_name}", dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
